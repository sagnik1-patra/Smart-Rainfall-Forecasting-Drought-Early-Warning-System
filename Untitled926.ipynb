```python
import os
import json
import yaml
import pickle
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    accuracy_score,
    classification_report,
    confusion_matrix
)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout


# ============================================================
# Paths
# ============================================================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Smart Rainfall Forecasting & Drought Early Warning System\Daily_Rainfall_District_Beed.csv"

OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Rainfall Forecasting & Drought Early Warning System"

os.makedirs(OUTPUT_DIR, exist_ok=True)


# ============================================================
# Load Dataset
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.strip()

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())


# ============================================================
# Column Names
# ============================================================

date_col = "Date"
rainfall_col = "Daily Rainfall (m.m)"
progressive_col = "Progressive (m.m)"
district_col = "District"


# ============================================================
# Cleaning
# ============================================================

df[date_col] = pd.to_datetime(df[date_col], errors="coerce", dayfirst=True)

df[rainfall_col] = pd.to_numeric(df[rainfall_col], errors="coerce")
df[progressive_col] = pd.to_numeric(df[progressive_col], errors="coerce")

df = df.dropna(subset=[date_col])

df[rainfall_col] = df[rainfall_col].fillna(df[rainfall_col].median())
df[progressive_col] = df[progressive_col].fillna(df[progressive_col].median())

df = df.sort_values(date_col).reset_index(drop=True)


# ============================================================
# Feature Engineering
# ============================================================

df["Day"] = df[date_col].dt.day
df["Month"] = df[date_col].dt.month
df["Year"] = df[date_col].dt.year
df["DayOfWeek"] = df[date_col].dt.dayofweek
df["WeekOfYear"] = df[date_col].dt.isocalendar().week.astype(int)

df["Rainfall_Lag_1"] = df[rainfall_col].shift(1)
df["Rainfall_Lag_2"] = df[rainfall_col].shift(2)
df["Rainfall_Lag_3"] = df[rainfall_col].shift(3)

df["Rolling_3_Day_Avg"] = df[rainfall_col].rolling(window=3, min_periods=1).mean()
df["Rolling_7_Day_Avg"] = df[rainfall_col].rolling(window=7, min_periods=1).mean()

df["Progressive_Growth"] = df[progressive_col].diff()
df["Rainy_Day"] = np.where(df[rainfall_col] > 2, 1, 0)

df["Next_Day_Rainfall"] = df[rainfall_col].shift(-1)
df["Next_Day_Rainfall"] = df["Next_Day_Rainfall"].fillna(df[rainfall_col])

df = df.fillna(0)


# ============================================================
# Rainfall and Drought Categories
# ============================================================

def rainfall_category(x):
    if x < 2:
        return "Low_Rainfall"
    elif x <= 10:
        return "Moderate_Rainfall"
    else:
        return "Heavy_Rainfall"

def drought_risk(x):
    if x < 2:
        return "High_Drought_Risk"
    elif x <= 10:
        return "Medium_Drought_Risk"
    else:
        return "Low_Drought_Risk"

df["Rainfall_Category"] = df[rainfall_col].apply(rainfall_category)
df["Drought_Risk"] = df[rainfall_col].apply(drought_risk)


# ============================================================
# Encoding
# ============================================================

district_encoder = LabelEncoder()
risk_encoder = LabelEncoder()

df["District_Encoded"] = district_encoder.fit_transform(df[district_col])
df["Drought_Risk_Encoded"] = risk_encoder.fit_transform(df["Drought_Risk"])


# ============================================================
# Features and Targets
# ============================================================

features = [
    "District_Encoded",
    "Day",
    "Month",
    "Year",
    "DayOfWeek",
    "WeekOfYear",
    "Rainfall_Lag_1",
    "Rainfall_Lag_2",
    "Rainfall_Lag_3",
    "Rolling_3_Day_Avg",
    "Rolling_7_Day_Avg",
    progressive_col,
    "Progressive_Growth",
    "Rainy_Day"
]

X = df[features]
y_regression = df["Next_Day_Rainfall"]
y_classification = df["Drought_Risk_Encoded"]


# ============================================================
# Train-Test Split
# ============================================================

test_size = 0.25 if len(df) >= 8 else 0.2

class_counts = y_classification.value_counts()

if len(class_counts) > 1 and class_counts.min() >= 2:
    stratify_value = y_classification
else:
    stratify_value = None

X_train, X_test, y_reg_train, y_reg_test, y_cls_train, y_cls_test = train_test_split(
    X,
    y_regression,
    y_classification,
    test_size=test_size,
    random_state=42,
    stratify=stratify_value
)


# ============================================================
# Scaling
# ============================================================

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


# ============================================================
# Random Forest Regression
# ============================================================

rf_regressor = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_regressor.fit(X_train_scaled, y_reg_train)

reg_pred = rf_regressor.predict(X_test_scaled)

regression_results = {
    "MAE": float(mean_absolute_error(y_reg_test, reg_pred)),
    "RMSE": float(np.sqrt(mean_squared_error(y_reg_test, reg_pred))),
    "R2_Score": float(r2_score(y_reg_test, reg_pred)) if len(y_reg_test) > 1 else 0.0
}

print("\nRegression Results:")
print(regression_results)


# ============================================================
# Random Forest Classification
# ============================================================

rf_classifier = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_classifier.fit(X_train_scaled, y_cls_train)

cls_pred = rf_classifier.predict(X_test_scaled)

all_labels = list(range(len(risk_encoder.classes_)))

classification_results = {
    "Accuracy": float(accuracy_score(y_cls_test, cls_pred)),
    "Classification_Report": classification_report(
        y_cls_test,
        cls_pred,
        labels=all_labels,
        target_names=risk_encoder.classes_,
        output_dict=True,
        zero_division=0
    ),
    "Confusion_Matrix": confusion_matrix(
        y_cls_test,
        cls_pred,
        labels=all_labels
    ).tolist()
}

print("\nClassification Accuracy:", classification_results["Accuracy"])


# ============================================================
# Neural Network Regression Model
# ============================================================

nn_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(16, activation="relu"),
    Dense(1)
])

nn_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

history = nn_model.fit(
    X_train_scaled,
    y_reg_train,
    epochs=100,
    batch_size=2,
    verbose=1
)

nn_pred = nn_model.predict(X_test_scaled).flatten()

nn_results = {
    "MAE": float(mean_absolute_error(y_reg_test, nn_pred)),
    "RMSE": float(np.sqrt(mean_squared_error(y_reg_test, nn_pred))),
    "R2_Score": float(r2_score(y_reg_test, nn_pred)) if len(y_reg_test) > 1 else 0.0
}

print("\nNeural Network Results:")
print(nn_results)


# ============================================================
# Save Models
# ============================================================

h5_path = os.path.join(OUTPUT_DIR, "rainfall_neural_network_model.h5")
pkl_reg_path = os.path.join(OUTPUT_DIR, "rainfall_random_forest_regressor.pkl")
pkl_cls_path = os.path.join(OUTPUT_DIR, "drought_risk_random_forest_classifier.pkl")
scaler_path = os.path.join(OUTPUT_DIR, "rainfall_scaler.pkl")
encoder_path = os.path.join(OUTPUT_DIR, "rainfall_label_encoders.pkl")

nn_model.save(h5_path)

with open(pkl_reg_path, "wb") as f:
    pickle.dump(rf_regressor, f)

with open(pkl_cls_path, "wb") as f:
    pickle.dump(rf_classifier, f)

with open(scaler_path, "wb") as f:
    pickle.dump(scaler, f)

with open(encoder_path, "wb") as f:
    pickle.dump({
        "district_encoder": district_encoder,
        "risk_encoder": risk_encoder
    }, f)


# ============================================================
# Save YAML Config
# ============================================================

config = {
    "project_name": "Smart Rainfall Forecasting and Drought Early Warning System",
    "author": "Sagnik Patra",
    "dataset_path": DATA_PATH,
    "output_directory": OUTPUT_DIR,
    "features": features,
    "target_regression": "Next_Day_Rainfall",
    "target_classification": "Drought_Risk",
    "models": {
        "regression": "RandomForestRegressor",
        "classification": "RandomForestClassifier",
        "deep_learning": "Neural Network Regression"
    },
    "drought_risk_rules": {
        "High_Drought_Risk": "Rainfall < 2 mm",
        "Medium_Drought_Risk": "Rainfall 2 to 10 mm",
        "Low_Drought_Risk": "Rainfall > 10 mm"
    }
}

yaml_path = os.path.join(OUTPUT_DIR, "rainfall_project_config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(config, f, sort_keys=False)


# ============================================================
# Save JSON Results
# ============================================================

results = {
    "dataset_shape": list(df.shape),
    "class_names": risk_encoder.classes_.tolist(),
    "regression_results_random_forest": regression_results,
    "classification_results_random_forest": classification_results,
    "neural_network_regression_results": nn_results,
    "saved_files": {
        "h5_model": h5_path,
        "random_forest_regressor": pkl_reg_path,
        "random_forest_classifier": pkl_cls_path,
        "scaler": scaler_path,
        "encoders": encoder_path,
        "yaml_config": yaml_path
    }
}

json_path = os.path.join(OUTPUT_DIR, "rainfall_model_results.json")

with open(json_path, "w") as f:
    json.dump(results, f, indent=4)


# ============================================================
# Save Prediction CSV
# ============================================================

prediction_df = X_test.copy()

prediction_df["Actual_Next_Day_Rainfall"] = y_reg_test.values
prediction_df["RF_Predicted_Rainfall"] = reg_pred
prediction_df["NN_Predicted_Rainfall"] = nn_pred

prediction_df["Actual_Drought_Risk"] = risk_encoder.inverse_transform(y_cls_test.values)
prediction_df["Predicted_Drought_Risk"] = risk_encoder.inverse_transform(cls_pred)

prediction_csv_path = os.path.join(OUTPUT_DIR, "rainfall_drought_predictions.csv")
prediction_df.to_csv(prediction_csv_path, index=False)


# ============================================================
# Final Output
# ============================================================

print("\nAll files saved successfully in:")
print(OUTPUT_DIR)

print("\nGenerated Files:")
print("1.", h5_path)
print("2.", pkl_reg_path)
print("3.", pkl_cls_path)
print("4.", scaler_path)
print("5.", encoder_path)
print("6.", yaml_path)
print("7.", json_path)
print("8.", prediction_csv_path)
```
