```python
import os
import json
import yaml
import pickle
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    accuracy_score,
    classification_report,
    confusion_matrix
)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout


# ============================================================
# Paths
# ============================================================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Smart Rainfall Forecasting & Drought Early Warning System\Daily_Rainfall_District_Beed.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Rainfall Forecasting & Drought Early Warning System"

os.makedirs(OUTPUT_DIR, exist_ok=True)


# ============================================================
# Load Dataset
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.strip()

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())


# ============================================================
# Column Detection
# ============================================================

date_col = "Date"
rainfall_col = "Daily Rainfall (m.m)"
progressive_col = "Progressive (m.m)"
district_col = "District"


# ============================================================
# Cleaning
# ============================================================

df[date_col] = pd.to_datetime(df[date_col], errors="coerce", dayfirst=True)

df[rainfall_col] = pd.to_numeric(df[rainfall_col], errors="coerce")
df[progressive_col] = pd.to_numeric(df[progressive_col], errors="coerce")

df = df.dropna(subset=[date_col])
df[rainfall_col] = df[rainfall_col].fillna(df[rainfall_col].median())
df[progressive_col] = df[progressive_col].fillna(df[progressive_col].median())

df = df.sort_values(date_col).reset_index(drop=True)


# ============================================================
# Feature Engineering
# ============================================================

df["Day"] = df[date_col].dt.day
df["Month"] = df[date_col].dt.month
df["Year"] = df[date_col].dt.year
df["DayOfWeek"] = df[date_col].dt.dayofweek
df["WeekOfYear"] = df[date_col].dt.isocalendar().week.astype(int)

df["Rainfall_Lag_1"] = df[rainfall_col].shift(1)
df["Rainfall_Lag_2"] = df[rainfall_col].shift(2)
df["Rainfall_Lag_3"] = df[rainfall_col].shift(3)

df["Rolling_3_Day_Avg"] = df[rainfall_col].rolling(window=3, min_periods=1).mean()
df["Rolling_7_Day_Avg"] = df[rainfall_col].rolling(window=7, min_periods=1).mean()

df["Progressive_Growth"] = df[progressive_col].diff()
df["Rainy_Day"] = np.where(df[rainfall_col] > 2, 1, 0)

df["Next_Day_Rainfall"] = df[rainfall_col].shift(-1)
df["Next_Day_Rainfall"] = df["Next_Day_Rainfall"].fillna(df[rainfall_col])

df = df.fillna(0)


# ============================================================
# Drought Risk Label
# ============================================================

def drought_risk(x):
    if x < 2:
        return "High_Drought_Risk"
    elif x <= 10:
        return "Medium_Drought_Risk"
    else:
        return "Low_Drought_Risk"

df["Drought_Risk"] = df[rainfall_col].apply(drought_risk)


# ============================================================
# Encoding
# ============================================================

district_encoder = LabelEncoder()
risk_encoder = LabelEncoder()

df["District_Encoded"] = district_encoder.fit_transform(df[district_col])
df["Drought_Risk_Encoded"] = risk_encoder.fit_transform(df["Drought_Risk"])


# ============================================================
# Features and Targets
# ============================================================

all_features = [
    "District_Encoded",
    "Day",
    "Month",
    "Year",
    "DayOfWeek",
    "WeekOfYear",
    "Rainfall_Lag_1",
    "Rainfall_Lag_2",
    "Rainfall_Lag_3",
    "Rolling_3_Day_Avg",
    "Rolling_7_Day_Avg",
    progressive_col,
    "Progressive_Growth",
    "Rainy_Day"
]

X = df[all_features]
y_regression = df["Next_Day_Rainfall"]
y_classification = df["Drought_Risk_Encoded"]


# ============================================================
# Split
# ============================================================

class_counts = y_classification.value_counts()

if len(class_counts) > 1 and class_counts.min() >= 2:
    stratify_value = y_classification
else:
    stratify_value = None

X_train, X_test, y_reg_train, y_reg_test, y_cls_train, y_cls_test = train_test_split(
    X,
    y_regression,
    y_classification,
    test_size=0.25,
    random_state=42,
    stratify=stratify_value
)


# ============================================================
# Scaling
# ============================================================

scaler = StandardScaler()
X_train_scaled_full = scaler.fit_transform(X_train)
X_test_scaled_full = scaler.transform(X_test)


# ============================================================
# AIS Feature Selection
# Artificial Immune System / Clonal Selection Inspired
# ============================================================

np.random.seed(42)

num_features = X_train_scaled_full.shape[1]
population_size = 20
generations = 25
clone_factor = 5
mutation_rate = 0.15

def create_antibody():
    antibody = np.random.randint(0, 2, num_features)

    if antibody.sum() == 0:
        antibody[np.random.randint(0, num_features)] = 1

    return antibody

def fitness_function(antibody):
    selected_indices = np.where(antibody == 1)[0]

    if len(selected_indices) == 0:
        return -9999

    X_tr = X_train_scaled_full[:, selected_indices]
    X_te = X_test_scaled_full[:, selected_indices]

    model = RandomForestRegressor(
        n_estimators=50,
        random_state=42
    )

    model.fit(X_tr, y_reg_train)
    pred = model.predict(X_te)

    rmse = np.sqrt(mean_squared_error(y_reg_test, pred))

    feature_penalty = len(selected_indices) / num_features

    fitness = -rmse - (0.05 * feature_penalty)

    return fitness

def mutate_antibody(antibody):
    mutated = antibody.copy()

    for i in range(len(mutated)):
        if np.random.rand() < mutation_rate:
            mutated[i] = 1 - mutated[i]

    if mutated.sum() == 0:
        mutated[np.random.randint(0, num_features)] = 1

    return mutated

population = np.array([create_antibody() for _ in range(population_size)])

best_fitness_history = []
avg_fitness_history = []

for gen in range(generations):
    fitness_scores = np.array([fitness_function(ab) for ab in population])

    sorted_indices = np.argsort(fitness_scores)[::-1]
    population = population[sorted_indices]
    fitness_scores = fitness_scores[sorted_indices]

    best_fitness_history.append(fitness_scores[0])
    avg_fitness_history.append(np.mean(fitness_scores))

    best_antibodies = population[:5]

    clones = []

    for antibody in best_antibodies:
        for _ in range(clone_factor):
            clone = mutate_antibody(antibody)
            clones.append(clone)

    random_antibodies = [
        create_antibody()
        for _ in range(population_size - len(clones))
    ]

    population = np.array(clones + random_antibodies)

final_fitness_scores = np.array([fitness_function(ab) for ab in population])
best_index = np.argmax(final_fitness_scores)
best_antibody = population[best_index]

selected_feature_indices = np.where(best_antibody == 1)[0]
selected_features = [all_features[i] for i in selected_feature_indices]

print("\nAIS Selected Features:")
print(selected_features)


# ============================================================
# Save AIS Feature Selection CSV
# ============================================================

ais_feature_df = pd.DataFrame({
    "Feature": all_features,
    "Selected": best_antibody,
    "Status": ["Selected" if x == 1 else "Not Selected" for x in best_antibody]
})

ais_feature_selection_csv_path = os.path.join(
    OUTPUT_DIR,
    "ais_feature_selection.csv"
)

ais_feature_df.to_csv(ais_feature_selection_csv_path, index=False)


# ============================================================
# Selected Feature Dataset
# ============================================================

X_train_selected = X_train_scaled_full[:, selected_feature_indices]
X_test_selected = X_test_scaled_full[:, selected_feature_indices]


# ============================================================
# AIS Optimized Random Forest Regression
# ============================================================

ais_rf_regressor = RandomForestRegressor(
    n_estimators=150,
    random_state=42
)

ais_rf_regressor.fit(X_train_selected, y_reg_train)

ais_rf_reg_pred = ais_rf_regressor.predict(X_test_selected)

ais_rf_mae = mean_absolute_error(y_reg_test, ais_rf_reg_pred)
ais_rf_rmse = np.sqrt(mean_squared_error(y_reg_test, ais_rf_reg_pred))
ais_rf_r2 = r2_score(y_reg_test, ais_rf_reg_pred) if len(y_reg_test) > 1 else 0.0


# ============================================================
# AIS Optimized Random Forest Classification
# ============================================================

ais_rf_classifier = RandomForestClassifier(
    n_estimators=150,
    random_state=42
)

ais_rf_classifier.fit(X_train_selected, y_cls_train)

ais_cls_pred = ais_rf_classifier.predict(X_test_selected)

ais_accuracy = accuracy_score(y_cls_test, ais_cls_pred)

all_labels = list(range(len(risk_encoder.classes_)))

ais_classification_report = classification_report(
    y_cls_test,
    ais_cls_pred,
    labels=all_labels,
    target_names=risk_encoder.classes_,
    output_dict=True,
    zero_division=0
)

ais_cm = confusion_matrix(
    y_cls_test,
    ais_cls_pred,
    labels=all_labels
)


# ============================================================
# AIS Neural Network Regression
# ============================================================

ais_nn_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train_selected.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(16, activation="relu"),
    Dense(1)
])

ais_nn_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

ais_history = ais_nn_model.fit(
    X_train_selected,
    y_reg_train,
    validation_split=0.2,
    epochs=100,
    batch_size=2,
    verbose=1
)

ais_nn_pred = ais_nn_model.predict(X_test_selected).flatten()

ais_nn_mae = mean_absolute_error(y_reg_test, ais_nn_pred)
ais_nn_rmse = np.sqrt(mean_squared_error(y_reg_test, ais_nn_pred))
ais_nn_r2 = r2_score(y_reg_test, ais_nn_pred) if len(y_reg_test) > 1 else 0.0


# ============================================================
# Save Models with ais_ prefix
# ============================================================

ais_h5_path = os.path.join(OUTPUT_DIR, "ais_rainfall_neural_network_model.h5")
ais_reg_pkl_path = os.path.join(OUTPUT_DIR, "ais_rainfall_random_forest_regressor.pkl")
ais_cls_pkl_path = os.path.join(OUTPUT_DIR, "ais_drought_risk_random_forest_classifier.pkl")
ais_scaler_path = os.path.join(OUTPUT_DIR, "ais_rainfall_scaler.pkl")
ais_encoder_path = os.path.join(OUTPUT_DIR, "ais_rainfall_label_encoders.pkl")

ais_nn_model.save(ais_h5_path)

with open(ais_reg_pkl_path, "wb") as f:
    pickle.dump(ais_rf_regressor, f)

with open(ais_cls_pkl_path, "wb") as f:
    pickle.dump(ais_rf_classifier, f)

with open(ais_scaler_path, "wb") as f:
    pickle.dump(scaler, f)

with open(ais_encoder_path, "wb") as f:
    pickle.dump({
        "district_encoder": district_encoder,
        "risk_encoder": risk_encoder,
        "selected_features": selected_features,
        "selected_feature_indices": selected_feature_indices.tolist()
    }, f)


# ============================================================
# Save Results CSV
# ============================================================

ais_results_df = pd.DataFrame({
    "Model": [
        "AIS + Random Forest Regressor",
        "AIS + Neural Network Regressor",
        "AIS + Random Forest Classifier"
    ],
    "Task": [
        "Rainfall Prediction",
        "Rainfall Prediction",
        "Drought Risk Classification"
    ],
    "MAE": [
        ais_rf_mae,
        ais_nn_mae,
        np.nan
    ],
    "RMSE": [
        ais_rf_rmse,
        ais_nn_rmse,
        np.nan
    ],
    "R2_Score": [
        ais_rf_r2,
        ais_nn_r2,
        np.nan
    ],
    "Accuracy": [
        np.nan,
        np.nan,
        ais_accuracy
    ],
    "Selected_Feature_Count": [
        len(selected_features),
        len(selected_features),
        len(selected_features)
    ]
})

ais_result_csv_path = os.path.join(OUTPUT_DIR, "ais_model_results.csv")
ais_results_df.to_csv(ais_result_csv_path, index=False)

print("\nAIS Model Results:")
print(ais_results_df)


# ============================================================
# Save Prediction CSV
# ============================================================

ais_prediction_df = X_test.copy()
ais_prediction_df["Actual_Next_Day_Rainfall"] = y_reg_test.values
ais_prediction_df["AIS_RF_Predicted_Rainfall"] = ais_rf_reg_pred
ais_prediction_df["AIS_NN_Predicted_Rainfall"] = ais_nn_pred
ais_prediction_df["Actual_Drought_Risk"] = risk_encoder.inverse_transform(y_cls_test.values)
ais_prediction_df["AIS_Predicted_Drought_Risk"] = risk_encoder.inverse_transform(ais_cls_pred)

ais_prediction_csv_path = os.path.join(OUTPUT_DIR, "ais_predictions.csv")
ais_prediction_df.to_csv(ais_prediction_csv_path, index=False)

print("\nAIS Prediction Results:")
print(ais_prediction_df.head())


# ============================================================
# Save JSON Results
# ============================================================

ais_json_results = {
    "project_name": "AIS Smart Rainfall Forecasting and Drought Early Warning System",
    "algorithm": "Artificial Immune System Feature Selection",
    "selected_features": selected_features,
    "selected_feature_count": len(selected_features),
    "all_features": all_features,
    "ais_random_forest_regression": {
        "MAE": float(ais_rf_mae),
        "RMSE": float(ais_rf_rmse),
        "R2_Score": float(ais_rf_r2)
    },
    "ais_neural_network_regression": {
        "MAE": float(ais_nn_mae),
        "RMSE": float(ais_nn_rmse),
        "R2_Score": float(ais_nn_r2)
    },
    "ais_random_forest_classification": {
        "Accuracy": float(ais_accuracy),
        "Classification_Report": ais_classification_report,
        "Confusion_Matrix": ais_cm.tolist()
    }
}

ais_json_path = os.path.join(OUTPUT_DIR, "ais_results.json")

with open(ais_json_path, "w") as f:
    json.dump(ais_json_results, f, indent=4)


# ============================================================
# Save YAML Config
# ============================================================

ais_yaml_config = {
    "project_name": "AIS Smart Rainfall Forecasting and Drought Early Warning System",
    "author": "Sagnik Patra",
    "algorithm": "Artificial Immune System",
    "dataset_path": DATA_PATH,
    "output_directory": OUTPUT_DIR,
    "selected_features": selected_features,
    "models": [
        "AIS + RandomForestRegressor",
        "AIS + RandomForestClassifier",
        "AIS + NeuralNetworkRegressor"
    ],
    "saved_outputs": [
        "ais_feature_selection.csv",
        "ais_model_results.csv",
        "ais_predictions.csv",
        "ais_results.json",
        "ais_config.yaml",
        "ais_rainfall_neural_network_model.h5",
        "ais_rainfall_random_forest_regressor.pkl",
        "ais_drought_risk_random_forest_classifier.pkl",
        "ais_accuracy_graph.png",
        "ais_confusion_matrix_heatmap.png",
        "ais_comparison_graph.png",
        "ais_result_graph.png",
        "ais_prediction_graph.png",
        "ais_fitness_curve.png",
        "ais_feature_selection_graph.png",
        "ais_rainfall_trend_graph.png",
        "ais_monthly_rainfall_graph.png",
        "ais_correlation_heatmap.png"
    ]
}

ais_yaml_path = os.path.join(OUTPUT_DIR, "ais_config.yaml")

with open(ais_yaml_path, "w") as f:
    yaml.dump(ais_yaml_config, f, sort_keys=False)


# ============================================================
# Graph 1: AIS Accuracy Graph
# ============================================================

plt.figure(figsize=(7, 5))
plt.bar(["AIS + RF Classifier"], [ais_accuracy])
plt.title("AIS Classification Accuracy Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.grid(axis="y", alpha=0.3)

ais_accuracy_graph_path = os.path.join(OUTPUT_DIR, "ais_accuracy_graph.png")
plt.savefig(ais_accuracy_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 2: AIS Confusion Matrix Heatmap
# ============================================================

plt.figure(figsize=(7, 5))
plt.imshow(ais_cm, interpolation="nearest")
plt.title("AIS Confusion Matrix Heatmap")
plt.colorbar()

tick_marks = np.arange(len(risk_encoder.classes_))
plt.xticks(tick_marks, risk_encoder.classes_, rotation=45)
plt.yticks(tick_marks, risk_encoder.classes_)

for i in range(ais_cm.shape[0]):
    for j in range(ais_cm.shape[1]):
        plt.text(j, i, ais_cm[i, j], ha="center", va="center")

plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")

ais_heatmap_path = os.path.join(OUTPUT_DIR, "ais_confusion_matrix_heatmap.png")
plt.savefig(ais_heatmap_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 3: AIS Regression Comparison Graph
# ============================================================

ais_comparison_df = ais_results_df[ais_results_df["Task"] == "Rainfall Prediction"]

plt.figure(figsize=(8, 5))
plt.bar(ais_comparison_df["Model"], ais_comparison_df["RMSE"])
plt.title("AIS Regression Model Comparison Using RMSE")
plt.xlabel("Model")
plt.ylabel("RMSE")
plt.xticks(rotation=20)
plt.grid(axis="y", alpha=0.3)

ais_comparison_graph_path = os.path.join(OUTPUT_DIR, "ais_comparison_graph.png")
plt.savefig(ais_comparison_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 4: AIS Result Graph
# ============================================================

ais_metric_df = ais_comparison_df[["Model", "MAE", "RMSE", "R2_Score"]].set_index("Model")

ais_metric_df.plot(kind="bar", figsize=(9, 5))
plt.title("AIS Regression Result Graph")
plt.xlabel("Model")
plt.ylabel("Metric Value")
plt.xticks(rotation=20)
plt.grid(axis="y", alpha=0.3)

ais_result_graph_path = os.path.join(OUTPUT_DIR, "ais_result_graph.png")
plt.savefig(ais_result_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 5: AIS Prediction Graph
# ============================================================

plt.figure(figsize=(10, 5))
plt.plot(range(len(y_reg_test)), y_reg_test.values, marker="o", label="Actual Rainfall")
plt.plot(range(len(ais_rf_reg_pred)), ais_rf_reg_pred, marker="o", label="AIS RF Predicted")
plt.plot(range(len(ais_nn_pred)), ais_nn_pred, marker="o", label="AIS NN Predicted")

plt.title("AIS Actual vs Predicted Rainfall")
plt.xlabel("Test Sample")
plt.ylabel("Rainfall")
plt.legend()
plt.grid(alpha=0.3)

ais_prediction_graph_path = os.path.join(OUTPUT_DIR, "ais_prediction_graph.png")
plt.savefig(ais_prediction_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 6: AIS Fitness Curve
# ============================================================

plt.figure(figsize=(8, 5))
plt.plot(best_fitness_history, marker="o", label="Best Fitness")
plt.plot(avg_fitness_history, marker="o", label="Average Fitness")
plt.title("AIS Fitness Curve")
plt.xlabel("Generation")
plt.ylabel("Fitness")
plt.legend()
plt.grid(alpha=0.3)

ais_fitness_curve_path = os.path.join(OUTPUT_DIR, "ais_fitness_curve.png")
plt.savefig(ais_fitness_curve_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 7: AIS Feature Selection Graph
# ============================================================

plt.figure(figsize=(10, 5))
plt.bar(ais_feature_df["Feature"], ais_feature_df["Selected"])
plt.title("AIS Selected Features Graph")
plt.xlabel("Feature")
plt.ylabel("Selected Status")
plt.xticks(rotation=90)
plt.grid(axis="y", alpha=0.3)

ais_feature_graph_path = os.path.join(OUTPUT_DIR, "ais_feature_selection_graph.png")
plt.savefig(ais_feature_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 8: Rainfall Trend Graph
# ============================================================

plt.figure(figsize=(12, 5))
plt.plot(df[date_col], df[rainfall_col])
plt.title("AIS Daily Rainfall Trend")
plt.xlabel("Date")
plt.ylabel("Daily Rainfall")
plt.grid(alpha=0.3)

ais_rainfall_trend_graph_path = os.path.join(OUTPUT_DIR, "ais_rainfall_trend_graph.png")
plt.savefig(ais_rainfall_trend_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 9: Monthly Rainfall Graph
# ============================================================

monthly_rainfall = df.groupby("Month")[rainfall_col].sum()

plt.figure(figsize=(8, 5))
plt.bar(monthly_rainfall.index, monthly_rainfall.values)
plt.title("AIS Monthly Rainfall Analysis")
plt.xlabel("Month")
plt.ylabel("Total Rainfall")
plt.grid(axis="y", alpha=0.3)

ais_monthly_graph_path = os.path.join(OUTPUT_DIR, "ais_monthly_rainfall_graph.png")
plt.savefig(ais_monthly_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 10: Correlation Heatmap
# ============================================================

corr = df[all_features + [rainfall_col, "Next_Day_Rainfall"]].corr()

plt.figure(figsize=(12, 8))
plt.imshow(corr, interpolation="nearest")
plt.title("AIS Feature Correlation Heatmap")
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

ais_correlation_heatmap_path = os.path.join(OUTPUT_DIR, "ais_correlation_heatmap.png")
plt.savefig(ais_correlation_heatmap_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Final Output
# ============================================================

print("\nAll AIS results saved successfully in:")
print(OUTPUT_DIR)

print("\nGenerated AIS Files:")
print("1.", ais_h5_path)
print("2.", ais_reg_pkl_path)
print("3.", ais_cls_pkl_path)
print("4.", ais_scaler_path)
print("5.", ais_encoder_path)
print("6.", ais_feature_selection_csv_path)
print("7.", ais_result_csv_path)
print("8.", ais_prediction_csv_path)
print("9.", ais_json_path)
print("10.", ais_yaml_path)
print("11.", ais_accuracy_graph_path)
print("12.", ais_heatmap_path)
print("13.", ais_comparison_graph_path)
print("14.", ais_result_graph_path)
print("15.", ais_prediction_graph_path)
print("16.", ais_fitness_curve_path)
print("17.", ais_feature_graph_path)
print("18.", ais_rainfall_trend_graph_path)
print("19.", ais_monthly_graph_path)
print("20.", ais_correlation_heatmap_path)
```
