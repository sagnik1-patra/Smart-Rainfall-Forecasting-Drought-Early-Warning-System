```python
import os
import json
import yaml
import pickle
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    accuracy_score,
    classification_report,
    confusion_matrix
)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout


# ============================================================
# Paths
# ============================================================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Smart Rainfall Forecasting & Drought Early Warning System\Daily_Rainfall_District_Beed.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Rainfall Forecasting & Drought Early Warning System"

os.makedirs(OUTPUT_DIR, exist_ok=True)


# ============================================================
# Load Dataset
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.strip()

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())


# ============================================================
# Columns
# ============================================================

date_col = "Date"
rainfall_col = "Daily Rainfall (m.m)"
progressive_col = "Progressive (m.m)"
district_col = "District"


# ============================================================
# Cleaning
# ============================================================

df[date_col] = pd.to_datetime(df[date_col], errors="coerce", dayfirst=True)

df[rainfall_col] = pd.to_numeric(df[rainfall_col], errors="coerce")
df[progressive_col] = pd.to_numeric(df[progressive_col], errors="coerce")

df = df.dropna(subset=[date_col])

df[rainfall_col] = df[rainfall_col].fillna(df[rainfall_col].median())
df[progressive_col] = df[progressive_col].fillna(df[progressive_col].median())

df = df.sort_values(date_col).reset_index(drop=True)


# ============================================================
# Feature Engineering
# ============================================================

df["Day"] = df[date_col].dt.day
df["Month"] = df[date_col].dt.month
df["Year"] = df[date_col].dt.year
df["DayOfWeek"] = df[date_col].dt.dayofweek
df["WeekOfYear"] = df[date_col].dt.isocalendar().week.astype(int)

df["Rainfall_Lag_1"] = df[rainfall_col].shift(1)
df["Rainfall_Lag_2"] = df[rainfall_col].shift(2)
df["Rainfall_Lag_3"] = df[rainfall_col].shift(3)

df["Rolling_3_Day_Avg"] = df[rainfall_col].rolling(window=3, min_periods=1).mean()
df["Rolling_7_Day_Avg"] = df[rainfall_col].rolling(window=7, min_periods=1).mean()

df["Progressive_Growth"] = df[progressive_col].diff()
df["Rainy_Day"] = np.where(df[rainfall_col] > 2, 1, 0)

df["Next_Day_Rainfall"] = df[rainfall_col].shift(-1)
df["Next_Day_Rainfall"] = df["Next_Day_Rainfall"].fillna(df[rainfall_col])

df = df.fillna(0)


# ============================================================
# Drought Risk Category
# ============================================================

def drought_risk(x):
    if x < 2:
        return "High_Drought_Risk"
    elif x <= 10:
        return "Medium_Drought_Risk"
    else:
        return "Low_Drought_Risk"

df["Drought_Risk"] = df[rainfall_col].apply(drought_risk)


# ============================================================
# Encoding
# ============================================================

district_encoder = LabelEncoder()
risk_encoder = LabelEncoder()

df["District_Encoded"] = district_encoder.fit_transform(df[district_col])
df["Drought_Risk_Encoded"] = risk_encoder.fit_transform(df["Drought_Risk"])


# ============================================================
# Features and Targets
# ============================================================

features = [
    "District_Encoded",
    "Day",
    "Month",
    "Year",
    "DayOfWeek",
    "WeekOfYear",
    "Rainfall_Lag_1",
    "Rainfall_Lag_2",
    "Rainfall_Lag_3",
    "Rolling_3_Day_Avg",
    "Rolling_7_Day_Avg",
    progressive_col,
    "Progressive_Growth",
    "Rainy_Day"
]

X = df[features]
y_regression = df["Next_Day_Rainfall"]
y_classification = df["Drought_Risk_Encoded"]


# ============================================================
# Train-Test Split
# ============================================================

class_counts = y_classification.value_counts()

if len(class_counts) > 1 and class_counts.min() >= 2:
    stratify_value = y_classification
else:
    stratify_value = None

X_train, X_test, y_reg_train, y_reg_test, y_cls_train, y_cls_test = train_test_split(
    X,
    y_regression,
    y_classification,
    test_size=0.25,
    random_state=42,
    stratify=stratify_value
)


# ============================================================
# Scaling
# ============================================================

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)


# ============================================================
# Random Forest Regression
# ============================================================

rf_regressor = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_regressor.fit(X_train_scaled, y_reg_train)
rf_reg_pred = rf_regressor.predict(X_test_scaled)

rf_reg_mae = mean_absolute_error(y_reg_test, rf_reg_pred)
rf_reg_rmse = np.sqrt(mean_squared_error(y_reg_test, rf_reg_pred))
rf_reg_r2 = r2_score(y_reg_test, rf_reg_pred) if len(y_reg_test) > 1 else 0.0


# ============================================================
# Random Forest Classification
# ============================================================

rf_classifier = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

rf_classifier.fit(X_train_scaled, y_cls_train)
rf_cls_pred = rf_classifier.predict(X_test_scaled)

rf_cls_accuracy = accuracy_score(y_cls_test, rf_cls_pred)

all_labels = list(range(len(risk_encoder.classes_)))

cls_report = classification_report(
    y_cls_test,
    rf_cls_pred,
    labels=all_labels,
    target_names=risk_encoder.classes_,
    output_dict=True,
    zero_division=0
)

cm = confusion_matrix(
    y_cls_test,
    rf_cls_pred,
    labels=all_labels
)


# ============================================================
# Neural Network Regression Model
# ============================================================

nn_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train_scaled.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(16, activation="relu"),
    Dense(1)
])

nn_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

history = nn_model.fit(
    X_train_scaled,
    y_reg_train,
    validation_split=0.2,
    epochs=100,
    batch_size=2,
    verbose=1
)

nn_reg_pred = nn_model.predict(X_test_scaled).flatten()

nn_reg_mae = mean_absolute_error(y_reg_test, nn_reg_pred)
nn_reg_rmse = np.sqrt(mean_squared_error(y_reg_test, nn_reg_pred))
nn_reg_r2 = r2_score(y_reg_test, nn_reg_pred) if len(y_reg_test) > 1 else 0.0


# ============================================================
# Save Models
# ============================================================

h5_path = os.path.join(OUTPUT_DIR, "rainfall_neural_network_model.h5")
pkl_reg_path = os.path.join(OUTPUT_DIR, "rainfall_random_forest_regressor.pkl")
pkl_cls_path = os.path.join(OUTPUT_DIR, "drought_risk_random_forest_classifier.pkl")
scaler_path = os.path.join(OUTPUT_DIR, "rainfall_scaler.pkl")
encoder_path = os.path.join(OUTPUT_DIR, "rainfall_label_encoders.pkl")

nn_model.save(h5_path)

with open(pkl_reg_path, "wb") as f:
    pickle.dump(rf_regressor, f)

with open(pkl_cls_path, "wb") as f:
    pickle.dump(rf_classifier, f)

with open(scaler_path, "wb") as f:
    pickle.dump(scaler, f)

with open(encoder_path, "wb") as f:
    pickle.dump({
        "district_encoder": district_encoder,
        "risk_encoder": risk_encoder
    }, f)


# ============================================================
# Result CSV
# ============================================================

results_df = pd.DataFrame({
    "Model": [
        "Random Forest Regressor",
        "Neural Network Regressor",
        "Random Forest Classifier"
    ],
    "Task": [
        "Rainfall Prediction",
        "Rainfall Prediction",
        "Drought Risk Classification"
    ],
    "MAE": [
        rf_reg_mae,
        nn_reg_mae,
        np.nan
    ],
    "RMSE": [
        rf_reg_rmse,
        nn_reg_rmse,
        np.nan
    ],
    "R2_Score": [
        rf_reg_r2,
        nn_reg_r2,
        np.nan
    ],
    "Accuracy": [
        np.nan,
        np.nan,
        rf_cls_accuracy
    ]
})

result_csv_path = os.path.join(OUTPUT_DIR, "rainfall_model_results.csv")
results_df.to_csv(result_csv_path, index=False)

print("\nModel Results:")
print(results_df)


# ============================================================
# Prediction CSV
# ============================================================

prediction_df = X_test.copy()
prediction_df["Actual_Next_Day_Rainfall"] = y_reg_test.values
prediction_df["RF_Predicted_Rainfall"] = rf_reg_pred
prediction_df["NN_Predicted_Rainfall"] = nn_reg_pred
prediction_df["Actual_Drought_Risk"] = risk_encoder.inverse_transform(y_cls_test.values)
prediction_df["Predicted_Drought_Risk"] = risk_encoder.inverse_transform(rf_cls_pred)

prediction_csv_path = os.path.join(OUTPUT_DIR, "rainfall_drought_predictions.csv")
prediction_df.to_csv(prediction_csv_path, index=False)

print("\nPrediction Results:")
print(prediction_df.head())


# ============================================================
# Save JSON
# ============================================================

json_results = {
    "random_forest_regression": {
        "MAE": float(rf_reg_mae),
        "RMSE": float(rf_reg_rmse),
        "R2_Score": float(rf_reg_r2)
    },
    "neural_network_regression": {
        "MAE": float(nn_reg_mae),
        "RMSE": float(nn_reg_rmse),
        "R2_Score": float(nn_reg_r2)
    },
    "random_forest_classification": {
        "Accuracy": float(rf_cls_accuracy),
        "Classification_Report": cls_report,
        "Confusion_Matrix": cm.tolist()
    }
}

json_path = os.path.join(OUTPUT_DIR, "rainfall_model_results.json")

with open(json_path, "w") as f:
    json.dump(json_results, f, indent=4)


# ============================================================
# Save YAML Config
# ============================================================

yaml_config = {
    "project_name": "Smart Rainfall Forecasting and Drought Early Warning System",
    "author": "Sagnik Patra",
    "dataset_path": DATA_PATH,
    "output_directory": OUTPUT_DIR,
    "features": features,
    "models": [
        "RandomForestRegressor",
        "RandomForestClassifier",
        "NeuralNetworkRegressor"
    ],
    "saved_outputs": [
        "rainfall_model_results.csv",
        "rainfall_drought_predictions.csv",
        "rainfall_model_results.json",
        "rainfall_project_config.yaml",
        "rainfall_neural_network_model.h5",
        "rainfall_random_forest_regressor.pkl",
        "drought_risk_random_forest_classifier.pkl"
    ]
}

yaml_path = os.path.join(OUTPUT_DIR, "rainfall_project_config.yaml")

with open(yaml_path, "w") as f:
    yaml.dump(yaml_config, f, sort_keys=False)


# ============================================================
# Graph 1: Accuracy Graph
# ============================================================

plt.figure(figsize=(7, 5))
plt.bar(["Random Forest Classifier"], [rf_cls_accuracy])
plt.title("Classification Accuracy Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.grid(axis="y", alpha=0.3)

accuracy_graph_path = os.path.join(OUTPUT_DIR, "accuracy_graph.png")
plt.savefig(accuracy_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 2: Heatmap - Confusion Matrix
# ============================================================

plt.figure(figsize=(7, 5))
plt.imshow(cm, interpolation="nearest")
plt.title("Confusion Matrix Heatmap")
plt.colorbar()

tick_marks = np.arange(len(risk_encoder.classes_))
plt.xticks(tick_marks, risk_encoder.classes_, rotation=45)
plt.yticks(tick_marks, risk_encoder.classes_)

for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        plt.text(j, i, cm[i, j], ha="center", va="center")

plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")

heatmap_path = os.path.join(OUTPUT_DIR, "confusion_matrix_heatmap.png")
plt.savefig(heatmap_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 3: Regression Model Comparison Graph
# ============================================================

comparison_df = results_df[results_df["Task"] == "Rainfall Prediction"]

plt.figure(figsize=(8, 5))
plt.bar(comparison_df["Model"], comparison_df["RMSE"])
plt.title("Regression Model Comparison Using RMSE")
plt.xlabel("Model")
plt.ylabel("RMSE")
plt.xticks(rotation=20)
plt.grid(axis="y", alpha=0.3)

comparison_graph_path = os.path.join(OUTPUT_DIR, "comparison_graph.png")
plt.savefig(comparison_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 4: Result Graph - MAE, RMSE, R2
# ============================================================

metric_df = comparison_df[["Model", "MAE", "RMSE", "R2_Score"]].set_index("Model")

metric_df.plot(kind="bar", figsize=(9, 5))
plt.title("Regression Result Graph")
plt.xlabel("Model")
plt.ylabel("Metric Value")
plt.xticks(rotation=20)
plt.grid(axis="y", alpha=0.3)

result_graph_path = os.path.join(OUTPUT_DIR, "result_graph.png")
plt.savefig(result_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 5: Prediction Graph - Actual vs Predicted
# ============================================================

plt.figure(figsize=(10, 5))
plt.plot(range(len(y_reg_test)), y_reg_test.values, marker="o", label="Actual Rainfall")
plt.plot(range(len(rf_reg_pred)), rf_reg_pred, marker="o", label="RF Predicted Rainfall")
plt.plot(range(len(nn_reg_pred)), nn_reg_pred, marker="o", label="NN Predicted Rainfall")

plt.title("Actual vs Predicted Rainfall")
plt.xlabel("Test Sample")
plt.ylabel("Rainfall")
plt.legend()
plt.grid(alpha=0.3)

prediction_graph_path = os.path.join(OUTPUT_DIR, "prediction_graph.png")
plt.savefig(prediction_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 6: Rainfall Trend Graph
# ============================================================

plt.figure(figsize=(12, 5))
plt.plot(df[date_col], df[rainfall_col])
plt.title("Daily Rainfall Trend")
plt.xlabel("Date")
plt.ylabel("Daily Rainfall")
plt.grid(alpha=0.3)

rainfall_trend_graph_path = os.path.join(OUTPUT_DIR, "rainfall_trend_graph.png")
plt.savefig(rainfall_trend_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 7: Monthly Rainfall Graph
# ============================================================

monthly_rainfall = df.groupby("Month")[rainfall_col].sum()

plt.figure(figsize=(8, 5))
plt.bar(monthly_rainfall.index, monthly_rainfall.values)
plt.title("Monthly Rainfall Analysis")
plt.xlabel("Month")
plt.ylabel("Total Rainfall")
plt.grid(axis="y", alpha=0.3)

monthly_graph_path = os.path.join(OUTPUT_DIR, "monthly_rainfall_graph.png")
plt.savefig(monthly_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 8: Correlation Heatmap
# ============================================================

corr = df[features + [rainfall_col, "Next_Day_Rainfall"]].corr()

plt.figure(figsize=(12, 8))
plt.imshow(corr, interpolation="nearest")
plt.title("Feature Correlation Heatmap")
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

correlation_heatmap_path = os.path.join(OUTPUT_DIR, "correlation_heatmap.png")
plt.savefig(correlation_heatmap_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Final Output
# ============================================================

print("\nAll results saved successfully in:")
print(OUTPUT_DIR)

print("\nSaved Files:")
print("1.", h5_path)
print("2.", pkl_reg_path)
print("3.", pkl_cls_path)
print("4.", scaler_path)
print("5.", encoder_path)
print("6.", result_csv_path)
print("7.", prediction_csv_path)
print("8.", json_path)
print("9.", yaml_path)
print("10.", accuracy_graph_path)
print("11.", heatmap_path)
print("12.", comparison_graph_path)
print("13.", result_graph_path)
print("14.", prediction_graph_path)
print("15.", rainfall_trend_graph_path)
print("16.", monthly_graph_path)
print("17.", correlation_heatmap_path)
```
