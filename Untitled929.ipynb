```python
import os
import json
import yaml
import pickle
import warnings
warnings.filterwarnings("ignore")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestRegressor, RandomForestClassifier
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score,
    accuracy_score,
    classification_report,
    confusion_matrix
)

import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout


# ============================================================
# Paths
# ============================================================

DATA_PATH = r"C:\Users\NXTWAVE\Downloads\Smart Rainfall Forecasting & Drought Early Warning System\Daily_Rainfall_District_Beed.csv"
OUTPUT_DIR = r"C:\Users\NXTWAVE\Downloads\Smart Rainfall Forecasting & Drought Early Warning System"

os.makedirs(OUTPUT_DIR, exist_ok=True)


# ============================================================
# Load Dataset
# ============================================================

df = pd.read_csv(DATA_PATH)
df.columns = df.columns.str.strip()

print("Dataset Loaded Successfully")
print("Shape:", df.shape)
print("Columns:", df.columns.tolist())


# ============================================================
# Column Names
# ============================================================

date_col = "Date"
rainfall_col = "Daily Rainfall (m.m)"
progressive_col = "Progressive (m.m)"
district_col = "District"


# ============================================================
# Cleaning
# ============================================================

df[date_col] = pd.to_datetime(df[date_col], errors="coerce", dayfirst=True)

df[rainfall_col] = pd.to_numeric(df[rainfall_col], errors="coerce")
df[progressive_col] = pd.to_numeric(df[progressive_col], errors="coerce")

df = df.dropna(subset=[date_col])

df[rainfall_col] = df[rainfall_col].fillna(df[rainfall_col].median())
df[progressive_col] = df[progressive_col].fillna(df[progressive_col].median())

df = df.sort_values(date_col).reset_index(drop=True)


# ============================================================
# Feature Engineering
# ============================================================

df["Day"] = df[date_col].dt.day
df["Month"] = df[date_col].dt.month
df["Year"] = df[date_col].dt.year
df["DayOfWeek"] = df[date_col].dt.dayofweek
df["WeekOfYear"] = df[date_col].dt.isocalendar().week.astype(int)

df["Rainfall_Lag_1"] = df[rainfall_col].shift(1)
df["Rainfall_Lag_2"] = df[rainfall_col].shift(2)
df["Rainfall_Lag_3"] = df[rainfall_col].shift(3)

df["Rolling_3_Day_Avg"] = df[rainfall_col].rolling(window=3, min_periods=1).mean()
df["Rolling_7_Day_Avg"] = df[rainfall_col].rolling(window=7, min_periods=1).mean()

df["Progressive_Growth"] = df[progressive_col].diff()
df["Rainy_Day"] = np.where(df[rainfall_col] > 2, 1, 0)

df["Next_Day_Rainfall"] = df[rainfall_col].shift(-1)
df["Next_Day_Rainfall"] = df["Next_Day_Rainfall"].fillna(df[rainfall_col])

df = df.fillna(0)


# ============================================================
# Drought Risk Category
# ============================================================

def drought_risk(x):
    if x < 2:
        return "High_Drought_Risk"
    elif x <= 10:
        return "Medium_Drought_Risk"
    else:
        return "Low_Drought_Risk"

df["Drought_Risk"] = df[rainfall_col].apply(drought_risk)


# ============================================================
# Encoding
# ============================================================

district_encoder = LabelEncoder()
risk_encoder = LabelEncoder()

df["District_Encoded"] = district_encoder.fit_transform(df[district_col])
df["Drought_Risk_Encoded"] = risk_encoder.fit_transform(df["Drought_Risk"])


# ============================================================
# Features and Targets
# ============================================================

all_features = [
    "District_Encoded",
    "Day",
    "Month",
    "Year",
    "DayOfWeek",
    "WeekOfYear",
    "Rainfall_Lag_1",
    "Rainfall_Lag_2",
    "Rainfall_Lag_3",
    "Rolling_3_Day_Avg",
    "Rolling_7_Day_Avg",
    progressive_col,
    "Progressive_Growth",
    "Rainy_Day"
]

X = df[all_features]
y_regression = df["Next_Day_Rainfall"]
y_classification = df["Drought_Risk_Encoded"]


# ============================================================
# Train-Test Split
# ============================================================

class_counts = y_classification.value_counts()

if len(class_counts) > 1 and class_counts.min() >= 2:
    stratify_value = y_classification
else:
    stratify_value = None

X_train, X_test, y_reg_train, y_reg_test, y_cls_train, y_cls_test = train_test_split(
    X,
    y_regression,
    y_classification,
    test_size=0.25,
    random_state=42,
    stratify=stratify_value
)


# ============================================================
# Scaling
# ============================================================

scaler = StandardScaler()

X_train_scaled_full = scaler.fit_transform(X_train)
X_test_scaled_full = scaler.transform(X_test)


# ============================================================
# CSA Feature Selection
# Crow Search Algorithm Inspired
# ============================================================

np.random.seed(42)

num_features = X_train_scaled_full.shape[1]

num_crows = 25
iterations = 30
flight_length = 2.0
awareness_probability = 0.1

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def binary_position(position):
    prob = sigmoid(position)
    binary = np.where(prob >= 0.5, 1, 0)

    if binary.sum() == 0:
        binary[np.random.randint(0, num_features)] = 1

    return binary

def fitness_function(binary_solution):
    selected_indices = np.where(binary_solution == 1)[0]

    if len(selected_indices) == 0:
        return -9999

    X_tr = X_train_scaled_full[:, selected_indices]
    X_te = X_test_scaled_full[:, selected_indices]

    model = RandomForestRegressor(
        n_estimators=50,
        random_state=42
    )

    model.fit(X_tr, y_reg_train)
    pred = model.predict(X_te)

    rmse = np.sqrt(mean_squared_error(y_reg_test, pred))

    feature_penalty = len(selected_indices) / num_features

    fitness = -rmse - (0.05 * feature_penalty)

    return fitness

positions = np.random.uniform(-2, 2, size=(num_crows, num_features))
memories = positions.copy()

memory_binary = np.array([binary_position(pos) for pos in memories])
memory_fitness = np.array([fitness_function(sol) for sol in memory_binary])

best_fitness_history = []
avg_fitness_history = []

for iteration in range(iterations):

    new_positions = positions.copy()

    for i in range(num_crows):

        random_crow = np.random.randint(0, num_crows)

        if np.random.rand() > awareness_probability:
            r = np.random.rand()
            new_positions[i] = positions[i] + flight_length * r * (
                memories[random_crow] - positions[i]
            )
        else:
            new_positions[i] = np.random.uniform(-2, 2, size=num_features)

    positions = new_positions

    current_binary = np.array([binary_position(pos) for pos in positions])
    current_fitness = np.array([fitness_function(sol) for sol in current_binary])

    for i in range(num_crows):
        if current_fitness[i] > memory_fitness[i]:
            memories[i] = positions[i]
            memory_fitness[i] = current_fitness[i]

    best_fitness_history.append(np.max(memory_fitness))
    avg_fitness_history.append(np.mean(memory_fitness))

best_index = np.argmax(memory_fitness)
best_position = memories[best_index]
best_binary_solution = binary_position(best_position)

selected_feature_indices = np.where(best_binary_solution == 1)[0]
selected_features = [all_features[i] for i in selected_feature_indices]

print("\nCSA Selected Features:")
print(selected_features)


# ============================================================
# Save CSA Feature Selection CSV
# ============================================================

csa_feature_df = pd.DataFrame({
    "Feature": all_features,
    "Selected": best_binary_solution,
    "Status": [
        "Selected" if x == 1 else "Not Selected"
        for x in best_binary_solution
    ]
})

csa_feature_selection_csv_path = os.path.join(
    OUTPUT_DIR,
    "csa_feature_selection.csv"
)

csa_feature_df.to_csv(csa_feature_selection_csv_path, index=False)


# ============================================================
# Selected Dataset
# ============================================================

X_train_selected = X_train_scaled_full[:, selected_feature_indices]
X_test_selected = X_test_scaled_full[:, selected_feature_indices]


# ============================================================
# CSA Optimized Random Forest Regression
# ============================================================

csa_rf_regressor = RandomForestRegressor(
    n_estimators=150,
    random_state=42
)

csa_rf_regressor.fit(X_train_selected, y_reg_train)

csa_rf_reg_pred = csa_rf_regressor.predict(X_test_selected)

csa_rf_mae = mean_absolute_error(y_reg_test, csa_rf_reg_pred)
csa_rf_rmse = np.sqrt(mean_squared_error(y_reg_test, csa_rf_reg_pred))
csa_rf_r2 = r2_score(y_reg_test, csa_rf_reg_pred) if len(y_reg_test) > 1 else 0.0


# ============================================================
# CSA Optimized Random Forest Classification
# ============================================================

csa_rf_classifier = RandomForestClassifier(
    n_estimators=150,
    random_state=42
)

csa_rf_classifier.fit(X_train_selected, y_cls_train)

csa_cls_pred = csa_rf_classifier.predict(X_test_selected)

csa_accuracy = accuracy_score(y_cls_test, csa_cls_pred)

all_labels = list(range(len(risk_encoder.classes_)))

csa_classification_report = classification_report(
    y_cls_test,
    csa_cls_pred,
    labels=all_labels,
    target_names=risk_encoder.classes_,
    output_dict=True,
    zero_division=0
)

csa_cm = confusion_matrix(
    y_cls_test,
    csa_cls_pred,
    labels=all_labels
)


# ============================================================
# CSA Neural Network Regression
# ============================================================

csa_nn_model = Sequential([
    Dense(64, activation="relu", input_shape=(X_train_selected.shape[1],)),
    Dropout(0.2),
    Dense(32, activation="relu"),
    Dropout(0.2),
    Dense(16, activation="relu"),
    Dense(1)
])

csa_nn_model.compile(
    optimizer="adam",
    loss="mse",
    metrics=["mae"]
)

csa_history = csa_nn_model.fit(
    X_train_selected,
    y_reg_train,
    validation_split=0.2,
    epochs=100,
    batch_size=2,
    verbose=1
)

csa_nn_pred = csa_nn_model.predict(X_test_selected).flatten()

csa_nn_mae = mean_absolute_error(y_reg_test, csa_nn_pred)
csa_nn_rmse = np.sqrt(mean_squared_error(y_reg_test, csa_nn_pred))
csa_nn_r2 = r2_score(y_reg_test, csa_nn_pred) if len(y_reg_test) > 1 else 0.0


# ============================================================
# Save Models with csa_ prefix
# ============================================================

csa_h5_path = os.path.join(
    OUTPUT_DIR,
    "csa_rainfall_neural_network_model.h5"
)

csa_reg_pkl_path = os.path.join(
    OUTPUT_DIR,
    "csa_rainfall_random_forest_regressor.pkl"
)

csa_cls_pkl_path = os.path.join(
    OUTPUT_DIR,
    "csa_drought_risk_random_forest_classifier.pkl"
)

csa_scaler_path = os.path.join(
    OUTPUT_DIR,
    "csa_rainfall_scaler.pkl"
)

csa_encoder_path = os.path.join(
    OUTPUT_DIR,
    "csa_rainfall_label_encoders.pkl"
)

csa_nn_model.save(csa_h5_path)

with open(csa_reg_pkl_path, "wb") as f:
    pickle.dump(csa_rf_regressor, f)

with open(csa_cls_pkl_path, "wb") as f:
    pickle.dump(csa_rf_classifier, f)

with open(csa_scaler_path, "wb") as f:
    pickle.dump(scaler, f)

with open(csa_encoder_path, "wb") as f:
    pickle.dump({
        "district_encoder": district_encoder,
        "risk_encoder": risk_encoder,
        "selected_features": selected_features,
        "selected_feature_indices": selected_feature_indices.tolist()
    }, f)


# ============================================================
# Save CSA Results CSV
# ============================================================

csa_results_df = pd.DataFrame({
    "Model": [
        "CSA + Random Forest Regressor",
        "CSA + Neural Network Regressor",
        "CSA + Random Forest Classifier"
    ],
    "Task": [
        "Rainfall Prediction",
        "Rainfall Prediction",
        "Drought Risk Classification"
    ],
    "MAE": [
        csa_rf_mae,
        csa_nn_mae,
        np.nan
    ],
    "RMSE": [
        csa_rf_rmse,
        csa_nn_rmse,
        np.nan
    ],
    "R2_Score": [
        csa_rf_r2,
        csa_nn_r2,
        np.nan
    ],
    "Accuracy": [
        np.nan,
        np.nan,
        csa_accuracy
    ],
    "Selected_Feature_Count": [
        len(selected_features),
        len(selected_features),
        len(selected_features)
    ]
})

csa_result_csv_path = os.path.join(
    OUTPUT_DIR,
    "csa_model_results.csv"
)

csa_results_df.to_csv(csa_result_csv_path, index=False)

print("\nCSA Model Results:")
print(csa_results_df)


# ============================================================
# Save CSA Prediction CSV
# ============================================================

csa_prediction_df = X_test.copy()

csa_prediction_df["Actual_Next_Day_Rainfall"] = y_reg_test.values
csa_prediction_df["CSA_RF_Predicted_Rainfall"] = csa_rf_reg_pred
csa_prediction_df["CSA_NN_Predicted_Rainfall"] = csa_nn_pred
csa_prediction_df["Actual_Drought_Risk"] = risk_encoder.inverse_transform(y_cls_test.values)
csa_prediction_df["CSA_Predicted_Drought_Risk"] = risk_encoder.inverse_transform(csa_cls_pred)

csa_prediction_csv_path = os.path.join(
    OUTPUT_DIR,
    "csa_predictions.csv"
)

csa_prediction_df.to_csv(csa_prediction_csv_path, index=False)

print("\nCSA Prediction Results:")
print(csa_prediction_df.head())


# ============================================================
# Save CSA JSON
# ============================================================

csa_json_results = {
    "project_name": "CSA Smart Rainfall Forecasting and Drought Early Warning System",
    "algorithm": "Crow Search Algorithm Feature Selection",
    "selected_features": selected_features,
    "selected_feature_count": len(selected_features),
    "all_features": all_features,
    "csa_random_forest_regression": {
        "MAE": float(csa_rf_mae),
        "RMSE": float(csa_rf_rmse),
        "R2_Score": float(csa_rf_r2)
    },
    "csa_neural_network_regression": {
        "MAE": float(csa_nn_mae),
        "RMSE": float(csa_nn_rmse),
        "R2_Score": float(csa_nn_r2)
    },
    "csa_random_forest_classification": {
        "Accuracy": float(csa_accuracy),
        "Classification_Report": csa_classification_report,
        "Confusion_Matrix": csa_cm.tolist()
    }
}

csa_json_path = os.path.join(
    OUTPUT_DIR,
    "csa_results.json"
)

with open(csa_json_path, "w") as f:
    json.dump(csa_json_results, f, indent=4)


# ============================================================
# Save CSA YAML Config
# ============================================================

csa_yaml_config = {
    "project_name": "CSA Smart Rainfall Forecasting and Drought Early Warning System",
    "author": "Sagnik Patra",
    "algorithm": "Crow Search Algorithm",
    "dataset_path": DATA_PATH,
    "output_directory": OUTPUT_DIR,
    "selected_features": selected_features,
    "models": [
        "CSA + RandomForestRegressor",
        "CSA + RandomForestClassifier",
        "CSA + NeuralNetworkRegressor"
    ],
    "saved_outputs": [
        "csa_feature_selection.csv",
        "csa_model_results.csv",
        "csa_predictions.csv",
        "csa_results.json",
        "csa_config.yaml",
        "csa_rainfall_neural_network_model.h5",
        "csa_rainfall_random_forest_regressor.pkl",
        "csa_drought_risk_random_forest_classifier.pkl",
        "csa_accuracy_graph.png",
        "csa_confusion_matrix_heatmap.png",
        "csa_comparison_graph.png",
        "csa_result_graph.png",
        "csa_prediction_graph.png",
        "csa_fitness_curve.png",
        "csa_feature_selection_graph.png",
        "csa_rainfall_trend_graph.png",
        "csa_monthly_rainfall_graph.png",
        "csa_correlation_heatmap.png"
    ]
}

csa_yaml_path = os.path.join(
    OUTPUT_DIR,
    "csa_config.yaml"
)

with open(csa_yaml_path, "w") as f:
    yaml.dump(csa_yaml_config, f, sort_keys=False)


# ============================================================
# Graph 1: CSA Accuracy Graph
# ============================================================

plt.figure(figsize=(7, 5))
plt.bar(["CSA + RF Classifier"], [csa_accuracy])
plt.title("CSA Classification Accuracy Graph")
plt.xlabel("Model")
plt.ylabel("Accuracy")
plt.ylim(0, 1)
plt.grid(axis="y", alpha=0.3)

csa_accuracy_graph_path = os.path.join(
    OUTPUT_DIR,
    "csa_accuracy_graph.png"
)

plt.savefig(csa_accuracy_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 2: CSA Confusion Matrix Heatmap
# ============================================================

plt.figure(figsize=(7, 5))
plt.imshow(csa_cm, interpolation="nearest")
plt.title("CSA Confusion Matrix Heatmap")
plt.colorbar()

tick_marks = np.arange(len(risk_encoder.classes_))
plt.xticks(tick_marks, risk_encoder.classes_, rotation=45)
plt.yticks(tick_marks, risk_encoder.classes_)

for i in range(csa_cm.shape[0]):
    for j in range(csa_cm.shape[1]):
        plt.text(j, i, csa_cm[i, j], ha="center", va="center")

plt.xlabel("Predicted Label")
plt.ylabel("Actual Label")

csa_heatmap_path = os.path.join(
    OUTPUT_DIR,
    "csa_confusion_matrix_heatmap.png"
)

plt.savefig(csa_heatmap_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 3: CSA Regression Comparison Graph
# ============================================================

csa_comparison_df = csa_results_df[
    csa_results_df["Task"] == "Rainfall Prediction"
]

plt.figure(figsize=(8, 5))
plt.bar(csa_comparison_df["Model"], csa_comparison_df["RMSE"])
plt.title("CSA Regression Model Comparison Using RMSE")
plt.xlabel("Model")
plt.ylabel("RMSE")
plt.xticks(rotation=20)
plt.grid(axis="y", alpha=0.3)

csa_comparison_graph_path = os.path.join(
    OUTPUT_DIR,
    "csa_comparison_graph.png"
)

plt.savefig(csa_comparison_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 4: CSA Result Graph
# ============================================================

csa_metric_df = csa_comparison_df[
    ["Model", "MAE", "RMSE", "R2_Score"]
].set_index("Model")

csa_metric_df.plot(kind="bar", figsize=(9, 5))
plt.title("CSA Regression Result Graph")
plt.xlabel("Model")
plt.ylabel("Metric Value")
plt.xticks(rotation=20)
plt.grid(axis="y", alpha=0.3)

csa_result_graph_path = os.path.join(
    OUTPUT_DIR,
    "csa_result_graph.png"
)

plt.savefig(csa_result_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 5: CSA Prediction Graph
# ============================================================

plt.figure(figsize=(10, 5))
plt.plot(range(len(y_reg_test)), y_reg_test.values, marker="o", label="Actual Rainfall")
plt.plot(range(len(csa_rf_reg_pred)), csa_rf_reg_pred, marker="o", label="CSA RF Predicted")
plt.plot(range(len(csa_nn_pred)), csa_nn_pred, marker="o", label="CSA NN Predicted")

plt.title("CSA Actual vs Predicted Rainfall")
plt.xlabel("Test Sample")
plt.ylabel("Rainfall")
plt.legend()
plt.grid(alpha=0.3)

csa_prediction_graph_path = os.path.join(
    OUTPUT_DIR,
    "csa_prediction_graph.png"
)

plt.savefig(csa_prediction_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 6: CSA Fitness Curve
# ============================================================

plt.figure(figsize=(8, 5))
plt.plot(best_fitness_history, marker="o", label="Best Fitness")
plt.plot(avg_fitness_history, marker="o", label="Average Fitness")
plt.title("CSA Fitness Curve")
plt.xlabel("Iteration")
plt.ylabel("Fitness")
plt.legend()
plt.grid(alpha=0.3)

csa_fitness_curve_path = os.path.join(
    OUTPUT_DIR,
    "csa_fitness_curve.png"
)

plt.savefig(csa_fitness_curve_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 7: CSA Feature Selection Graph
# ============================================================

plt.figure(figsize=(10, 5))
plt.bar(csa_feature_df["Feature"], csa_feature_df["Selected"])
plt.title("CSA Selected Features Graph")
plt.xlabel("Feature")
plt.ylabel("Selected Status")
plt.xticks(rotation=90)
plt.grid(axis="y", alpha=0.3)

csa_feature_graph_path = os.path.join(
    OUTPUT_DIR,
    "csa_feature_selection_graph.png"
)

plt.savefig(csa_feature_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 8: Rainfall Trend Graph
# ============================================================

plt.figure(figsize=(12, 5))
plt.plot(df[date_col], df[rainfall_col])
plt.title("CSA Daily Rainfall Trend")
plt.xlabel("Date")
plt.ylabel("Daily Rainfall")
plt.grid(alpha=0.3)

csa_rainfall_trend_graph_path = os.path.join(
    OUTPUT_DIR,
    "csa_rainfall_trend_graph.png"
)

plt.savefig(csa_rainfall_trend_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 9: Monthly Rainfall Graph
# ============================================================

monthly_rainfall = df.groupby("Month")[rainfall_col].sum()

plt.figure(figsize=(8, 5))
plt.bar(monthly_rainfall.index, monthly_rainfall.values)
plt.title("CSA Monthly Rainfall Analysis")
plt.xlabel("Month")
plt.ylabel("Total Rainfall")
plt.grid(axis="y", alpha=0.3)

csa_monthly_graph_path = os.path.join(
    OUTPUT_DIR,
    "csa_monthly_rainfall_graph.png"
)

plt.savefig(csa_monthly_graph_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Graph 10: Correlation Heatmap
# ============================================================

corr = df[all_features + [rainfall_col, "Next_Day_Rainfall"]].corr()

plt.figure(figsize=(12, 8))
plt.imshow(corr, interpolation="nearest")
plt.title("CSA Feature Correlation Heatmap")
plt.colorbar()

plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)

csa_correlation_heatmap_path = os.path.join(
    OUTPUT_DIR,
    "csa_correlation_heatmap.png"
)

plt.savefig(csa_correlation_heatmap_path, dpi=300, bbox_inches="tight")
plt.show()


# ============================================================
# Final Output
# ============================================================

print("\nAll CSA results saved successfully in:")
print(OUTPUT_DIR)

print("\nGenerated CSA Files:")
print("1.", csa_h5_path)
print("2.", csa_reg_pkl_path)
print("3.", csa_cls_pkl_path)
print("4.", csa_scaler_path)
print("5.", csa_encoder_path)
print("6.", csa_feature_selection_csv_path)
print("7.", csa_result_csv_path)
print("8.", csa_prediction_csv_path)
print("9.", csa_json_path)
print("10.", csa_yaml_path)
print("11.", csa_accuracy_graph_path)
print("12.", csa_heatmap_path)
print("13.", csa_comparison_graph_path)
print("14.", csa_result_graph_path)
print("15.", csa_prediction_graph_path)
print("16.", csa_fitness_curve_path)
print("17.", csa_feature_graph_path)
print("18.", csa_rainfall_trend_graph_path)
print("19.", csa_monthly_graph_path)
print("20.", csa_correlation_heatmap_path)
```
